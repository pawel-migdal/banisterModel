```python
import pandas as pd
```

```python
def ef(data):
    hr = data['HeartRateAverage']
    pace = data['PowerAverage']
    return pace / hr

def ef2(data):
    hr = data['HeartRateAverage']
    pace = data['VelocityAverage']
    return pace / hr

data = pd.read_csv('workoutss2t.csv')

bike  = data[data['WorkoutType'] == 'Bike']
run   = data[data['WorkoutType'] == 'Run']

bike['EF'] = ef(bike)
run['EF'] = ef2(run)

bike.plot.scatter(x='WorkoutDay', y='EF')
run.plot.scatter(x='WorkoutDay', y='EF')
```
